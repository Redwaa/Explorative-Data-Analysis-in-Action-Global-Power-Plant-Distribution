> **Q1: How much energy was estimated to be produced in 2020 on each continent?**

```python
Energy_per_continent=gdf_world.groupby("continent")
Energy_per_continent_tab=Energy_per_continent.agg({'estimated_generation_gwh_2020':['sum']})
Energy_per_continent_tab
```

> **Q2: Looking at each continent: which country produces the most energy on their continent?**

```python
answer=Energy_per_continent_tab[Energy_per_continent_tab['estimated_generation_gwh_2020',
'sum']== (Energy_per_continent_tab['estimated_generation_gwh_2020',
'sum'].max())]
print(answer)
```
